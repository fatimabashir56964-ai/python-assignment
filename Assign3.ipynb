```python
#  Question : 01
num1 = float(input("Enter First number: "))
num2 = float(input("Enter Second number: "))
operator = input("Enter any operator of your choice (+, -, *, /, **): ")

if operator == "+":
    print("Result: ",num1 + num2)

elif operator == "-":
    print("Result: ",num1 - num2)

elif operator == "*":
    print("Result: ",num1 * num2)

elif operator == "/":
        print("Result: ",num1 / num2)

elif operator == "**":
    print("Result: ",num1 ** num2)  

else:
    print("Invalid operator")
```

```python
#  Question : 02
colors = ["red", "blue", 4, "purple", "green"]
for item in colors:
    if type(item) == int:
        print("Numeric value found: ",item)
        break;
    else:    
        print("No numeric value found")
```

```python
#  Question : 04
nums = {
    "num1" : 14,
    "num2" : 20,
    "num3" : 7,
    "num4" : 11
}

total = 0
for value in nums.values():
    total += value

print("Sum of numbers: ",total)    
```

```python
#  Question : 05
nums = [2, 8, 11, 32, 9, 7, 2]

for item in nums:
    if nums.count(item) > 1:
        print("Duplicate: ",item)
        break;
    else:
        print("No duplicate value")
```
